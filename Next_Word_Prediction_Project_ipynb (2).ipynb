# Data Preprocessing


#### Text Cleaning:Text Cleaning: Removing special characters, numbers, and punctuation.


```python
import re
```

```python
re_special_char = r"[^a-z\s?!.,]"
```

```python
def text_preprocessing(text):
  text = str(text)
  text = text.lower()
  text = re.sub(re_special_char, "", text)
  return text
```

**Stopword Removal (optional):** Removing frequently used words that do not add significant meaning.


```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]
    return " ".join(filtered_words)

tweets["Tweets"] = tweets["Tweets"].apply(remove_stopwords)
```
